```python
import pandas as pd
import chardet
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']
file_path = 'Dataset.csv'
detected_encoding = detect_encoding(file_path)
data = pd.read_csv(file_path, encoding=detected_encoding)
```

```python
data
```

```python
data['problem']=data['problem'].str.lower()
data['solution']=data['solution'].str.lower()
data.head()
```

##Removing puctuation

```python
import string
string.punctuation
```

```python
def remove_punctuations(text):
    punctuations = string.punctuation
    return text.translate(str.maketrans('', '', punctuations))
```

```python
data['problem'] = data['problem'].apply(lambda x: remove_punctuations(x))
data.head()
```

```python
data['solution'] = data['solution'].astype(str).apply(lambda x: remove_punctuations(x))
data.head()
```

##Removal of Stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in STOPWORDS])
```

```python
data['problem'] = data['problem'].apply(lambda x: remove_stopwords(x))
data['solution'] = data['solution'].astype(str).apply(lambda x: remove_stopwords(x))
data.head()
```

##Removal of Special characters

```python
import re
def remove_spl_chars(text):
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
    text = re.sub('\s+', ' ', text)
    return text
```

```python
data['problem'] = data['problem'].apply(lambda x: remove_spl_chars(x))
data['solution'] = data['solution'].astype(str).apply(lambda x: remove_spl_chars(x))
data.head()
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}
def lemmatize_words(text):
    pos_text = pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_text])
wordnet.NOUN
```

```python
data['problem'] = data['problem'].apply(lambda x: lemmatize_words(x))
data['solution'] = data['solution'].astype(str).apply(lambda x: lemmatize_words(x))
data.head()
```

##Removal of rare words

```python
from collections import Counter
word_count = Counter()
for text in data['problem']:
    for word in text.split():
        word_count[word] += 1
```

```python
word_count.most_common(10)
```

```python
RARE_WORDS = set(word for (word, wc) in word_count.most_common()[:-10:-1])
RARE_WORDS
```

```python
def remove_rare_words(text):
    return " ".join([word for word in text.split() if word not in RARE_WORDS])
```

```python
data['solution'] = data['solution'].apply(lambda x: remove_rare_words(x))
data.head()
```

## Idea Validator

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

problems = []
solutions = []

print("Enter problem statements to check feasibility.")
problem_input = input("Enter a problem statement:\n")
problems.append(problem_input)

print("\nEnter solutions for above problem statements.")
solution_input = input("Enter a solution:\n")
solutions.append(solution_input)

vectorizer = TfidfVectorizer()
problem_vectors = vectorizer.fit_transform(problems)
solution_vectors = vectorizer.transform(solutions)

similarity_scores = cosine_similarity(problem_vectors, solution_vectors)

threshold = 0.7
print("\n")

for i, problem in enumerate(problems):
    for j, solution in enumerate(solutions):
        if similarity_scores[i, j] > threshold:
            print(f"Feasible (Similarity Score: {similarity_scores[i, j]:.4f})\n")
        else:
            print(f"Infeasible (Similarity Score: {similarity_scores[i, j]:.4f})\n")
```

```python
def idea_validator(solutions):
    feasibility_score = 0

    # Criteria 1: Idea clarity and specificity
    if len(solutions) > 0:
        feasibility_score += 1
    else:
        print("Idea is not clearly defined. Please provide a more specific description.")

    # Criteria 2: Potential target audience or users
    target_audience = input("Who is the target audience for this idea? ")
    if len(target_audience) > 0:
        feasibility_score += 1
    else:
        print("Target audience is not defined. Consider identifying potential users.")

    # Criteria 3: Value proposition
    value_proposition = input("What is the value proposition of the idea? ")
    if len(value_proposition) > 0:
        feasibility_score += 1
    else:
        print("Value proposition is not clear. Define the unique value your idea provides.")


    # Display feasibility assessment
    print("\nFeasibility Assessment:")
    if feasibility_score >= 5:
        print("The idea appears to be feasible.")
    else:
        print("The idea may require further refinement. Consider addressing the highlighted areas.")

if __name__ == "__main__":
    # Get idea input from the user
    user_idea = solutions

    # Validate the idea
    idea_validator(user_idea)
```
